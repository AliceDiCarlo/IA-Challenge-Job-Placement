# Dummy variables

```python
data_train.head()
```

```python
# All binary values from string to numeral

data_train.loc[data_train['gender'] == 'M', 'gender'] = 1.0
data_train.loc[data_train['gender'] == 'F', 'gender'] = 0.0

data_train.loc[data_train['status'] == 'Placed', 'status'] = 1
data_train.loc[data_train['status'] == 'Not Placed', 'status'] = 0

data_train.loc[data_train['workex'] == 'Yes', 'workex'] = 1.0
data_train.loc[data_train['workex'] == 'No', 'workex'] = 0.0


categorical_var = ['ssc_b','hsc_b','hsc_s','degree_t','specialisation']

#  other values
for variable in categorical_var:
#     create columns
    dummies = pd.get_dummies(data_train[variable],prefix=variable)
#     update and drop original columns
    data_train = pd.concat([data_train,dummies],axis=1)
    data_train.drop([variable],axis=1,inplace=True)


data_train.head()
```

```python
# All binary values from string to numeral

data_test.loc[data_test['gender'] == 'M', 'gender'] = 1.0
data_test.loc[data_test['gender'] == 'F', 'gender'] = 0.0

data_test.loc[data_test['status'] == 'Placed', 'status'] = 1
data_test.loc[data_test['status'] == 'Not Placed', 'status'] = 0


data_test.loc[data_test['workex'] == 'Yes', 'workex'] = 1.0
data_test.loc[data_test['workex'] == 'No', 'workex'] = 0.0


categorical_var = ['ssc_b','hsc_b','hsc_s','degree_t','specialisation']

#  other values
for variable in categorical_var:
#     create columns
    dummies = pd.get_dummies(data_test[variable],prefix=variable)
#     update and drop original columns
    data_test = pd.concat([data_test,dummies],axis=1)
    data_test.drop([variable],axis=1,inplace=True)
```

# Classifier

```python
x = data_train.drop(['status','salary'], axis=1)
y = data_train['status'].astype(float)


x_test = data_test.drop(['status','salary'], axis=1)
y_test = data_test['status'].astype(float)


lr_model = LogisticRegression()
lr_model.fit(x,y)


rf_regressor = RandomForestRegressor(100, oob_score=True, n_jobs=-1, random_state=42)
rf_regressor.fit(x,y)
print('Score: ', rf_regressor.score(x,y))

rf_classifier = RandomForestClassifier(200, oob_score=True, n_jobs=-1, random_state=42)
rf_classifier.fit(x, y)

rf_classifier.score(x_test, y_test)
```
